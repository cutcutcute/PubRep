```python
import pandas as pd 
import seaborn as sb 
import numpy as np 
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import confusion_matrix 
import matplotlib.pyplot as plt



train_features = pd.read_csv('data/train_features.csv')
train_labels = pd.read_csv('data/train_labels.csv')

test_features = pd.read_csv('data/test_features.csv')
test_labels = pd.read_csv('data/test_labels.csv')


sb.countplot(x= 'Survived', data =train_labels)
```

```python
log_reg = LogisticRegression() 
log_reg.fit(train_features, train_labels.values.ravel())

score = log_reg.score(test_features, test_labels)
score
```

```python
survival = log_reg.predict(test_features)
print(f'OK {sum(survival != 0)}')
print(f'NOT OK {sum(survival == 0)}')
```

```python
conf_mx = confusion_matrix(test_labels, survival)

sb.heatmap(conf_mx, annot= True, fmt = '4.0f', cmap = 'crest')
plt.title('Confusin metrix', size = 12)
conf_mx
```
